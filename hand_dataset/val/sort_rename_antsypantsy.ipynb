```python
import os

folders = []
for i in range(0, 6):
    folders.append(str(i))
```

```python
for num_fol in folders:
    files = os.listdir(num_fol)
    for i, file in enumerate(files):
        old = os.path.join(num_fol, file)
        new = os.path.join(num_fol, f"{num_fol}_{i}.jpg")
        os.rename(old, new)
```
